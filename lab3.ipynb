```python
from PIL import Image
import os

# Load both images
photo = Image.open("photo1.jpg").convert("RGB")
screenshot = Image.open("screenshot1.png").convert("RGB")

# Save at 4 quality levels
for quality in [95, 75, 50, 20]:
    photo.save(f"photo_q{quality}.jpg", "JPEG", quality=quality)
    screenshot.save(f"screenshot_q{quality}.jpg", "JPEG", quality=quality)

# Save as PNG
photo.save("photo1.png")
screenshot.save("screenshot1.png")

print("All files saved.")
```

```python
files = [
    "photo_q95.jpg", "photo_q75.jpg", "photo_q50.jpg", "photo_q20.jpg", "photo1.png",
    "screenshot_q95.jpg", "screenshot_q75.jpg", "screenshot_q50.jpg", "screenshot_q20.jpg", "screenshot1.png"
]

for f in files:
    size = os.path.getsize(f)
    print(f"{f:30}: {size:,} bytes")
```

```python
import matplotlib.pyplot as plt

photo_q95 = Image.open("photo_q95.jpg")
photo_q50 = Image.open("photo_q50.jpg")
photo_q20 = Image.open("photo_q20.jpg")
photo_png = Image.open("photo1.png")

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
axes[0].imshow(photo_q95); axes[0].set_title("JPEG q=95"); axes[0].axis("off")
axes[1].imshow(photo_q50); axes[1].set_title("JPEG q=50"); axes[1].axis("off")
axes[2].imshow(photo_q20); axes[2].set_title("JPEG q=20"); axes[2].axis("off")
axes[3].imshow(photo_png); axes[3].set_title("PNG");       axes[3].axis("off")
plt.suptitle("Photo comparison")
plt.tight_layout()
plt.show()
```
